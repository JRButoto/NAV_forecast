```python
import pandas as pd
# import the webdriver library that will help us to create a driver to scrape a specific website
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# take the link
web_link="https://www.uttamis.co.tz/fund-performance"

# Create a Chrome WebDriver instance with specified options (e.g., headless mode)
driver = webdriver.Chrome(options=Options())

# Open the web link specified by the 'web_link' variable
driver.get(web_link)

# Maximize the browser window for better visibility and interaction
driver.maximize_window()

# Find and select a specific option element on the webpage using XPath
button = driver.find_element(By.XPATH, '//*[@id="data_table_length"]/label/select/option[4]')

# Click the selected option to perform an action on the webpage
button.click()
```

```python
# Find all elements matching a specific XPath pattern for the 'name_scheme' column
name_scheme = driver.find_elements(By.XPATH, "//tbody/tr/td[2]")

# Find all elements matching a specific XPath pattern for the 'net_asset_value' column
net_asset_value = driver.find_elements(By.XPATH, "//tbody/tr/td[3]")

# Find all elements matching a specific XPath pattern for the 'outstanding_no_of_units' column
outstanding_no_of_units = driver.find_elements(By.XPATH, "//tbody/tr/td[4]")

# Find all elements matching a specific XPath pattern for the 'nav_per_unit' column
nav_per_unit = driver.find_elements(By.XPATH, "//tbody/tr/td[5]")

# Find all elements matching a specific XPath pattern for the 'sale_price_per_unit' column
sale_price_per_unit = driver.find_elements(By.XPATH, "//tbody/tr/td[6]")

# Find all elements matching a specific XPath pattern for the 'repurchase_price_per_unit' column
repurchase_price_per_unit = driver.find_elements(By.XPATH, "//tbody/tr/td[7]")

# Find all elements matching a specific XPath pattern for the 'date_valued' column
date_valued = driver.find_elements(By.XPATH, "//tbody/tr/td[8]")
```

```python
# Extracting data from multiple web elements and organizing it into a list of dictionaries.
# Each dictionary represents a row of data with columns such as 'name_scheme,' 'net_asset_value,' 'outstanding_no_of_units,' 'nav_per_unit,' 'sale_price_per_unit,' 'repurchase_price_per_unit,' and 'date_valued.'
# The 'for' loop iterates through elements of corresponding columns, extracts their text content, and creates a dictionary for each row of data.
# Each dictionary is then appended to the 'data' list, resulting in a structured representation of the web table data.

data=[]

for i in range(len(name_scheme)):
    temp={"name_scheme":name_scheme[i].text,
         "net_asset_value":net_asset_value[i].text,
         "outstanding_no_of_units":outstanding_no_of_units[i].text,
         "nav_per_unit":nav_per_unit[i].text,
         "sale_price_per_unit":sale_price_per_unit[i].text,
         "repurchase_price_per_unit":repurchase_price_per_unit[i].text,
         "date_valued":date_valued[i].text}
    data.append(temp)
    
```

```python
# converting it to a dataframe and displaying it
df=pd.DataFrame(data)
df
```
