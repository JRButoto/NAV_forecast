```python
# Import the pandas library and alias it as 'pd' for data manipulation and analysis
import pandas as pd

# Import the matplotlib library for creating plots and visualizations
import matplotlib.pyplot as plt

# Import the seaborn library and alias it as 'sns' for enhanced data visualization
import seaborn as sns
```

```python
# Read data from a CSV file named 'Clean_Nav.csv' and store it in a DataFrame called 'data'
data = pd.read_csv('Clean_Nav.csv')

# Display the first few rows of the 'data' DataFrame to inspect the data and its structure
data.head()
```

```python
df_grouped = data.groupby('Scheme Name')[['Sale Price/Unit', 'Repurchase Price/Unit']].mean()
df_grouped.plot(kind='bar', stacked=False)
plt.xlabel('Fund Name')
plt.ylabel('Average Value')
plt.title('Average Fund Performance by Fund')
plt.show()
```

```python
df_grouped = data.groupby('Scheme Name')[['NAV']].mean()
df_grouped.plot(kind='bar', stacked=False)
plt.xlabel('Fund Name')
plt.ylabel('Average Value')
plt.title('Average Fund Performance by Fund')
plt.show()
```

```python
df_grouped = data.groupby('Scheme Name')[['Units']].mean()
df_grouped.plot(kind='bar', stacked=False)
plt.xlabel('Fund Name')
plt.ylabel('Average Value')
plt.title('Average Fund Performance by Fund')
plt.show()
```

```python
correlation_matrix = data[['Sale Price/Unit', 'Repurchase Price/Unit', 'Units', 'NAV']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
funds=['Umoja Fund','Watoto Fund', 'Wekeza Maisha Fund','Jikimu Fund','Liquid Fund','Bond Fund']
for i in funds:
    datat = data[data['Scheme Name'] == i]
    data.boxplot(column=['Sale Price/Unit'], by='Year')
    plt.xlabel('Year')
    plt.ylabel('Price')
    plt.title(f'Sale Price Distribution Over Years - {i}')
    plt.show()
```
