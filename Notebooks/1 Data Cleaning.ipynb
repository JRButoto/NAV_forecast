```python
#import important modules
import pandas as pd
import dtale as dt
```

```python
#load the dataset
filepath='NAV Dataset.csv'
data=pd.read_csv(filepath)
```

```python
# Checking the data
data.info()
```

```python
# Checking counts for each scheme
data['Scheme Name'].value_counts()
```

```python
# checking null values
data.isnull().sum()
```

```python
data.head()
```

```python
#Since Nav per unit is equal to sale price per unit
# Drop a column(Nav Per unit)

data.drop('Nav Per Unit',axis=1, inplace=True)


# Rename a column
data.rename(columns={'Sale Price per Unit': 'Sale Price/Unit'}, inplace=True)
data.rename(columns={'Repurchase Price/Unit': 'Repurchase Price/Unit'}, inplace=True)
data.rename(columns={'Outstanding Number of Units': 'Units'}, inplace=True)
data.rename(columns={'Net Asset Value': 'NAV'}, inplace=True)


data.head()
```

```python
# Remove commas from the 'Units' column to prepare it for conversion to a numeric data type
data['Units'] = data['Units'].str.replace(',', '')

# Remove commas from the 'NAV' (Net Asset Value) column to prepare it for conversion to a numeric data type
data['NAV'] = data['NAV'].str.replace(',', '')

# Convert the 'NAV' column to a floating-point (float) data type after removing commas
data['NAV'] = data['NAV'].astype(float)

# Convert the 'Units' column to a floating-point (float) data type after removing commas
data['Units'] = data['Units'].astype(float)

# Convert the 'Date Valued' column to datetime format for better time-based analysis
data['Date Valued'] = pd.to_datetime(data['Date Valued'])

# Display summary information about the DataFrame, including data types and non-null counts
data.info()
```

```python
data.head()
data['Year'] = data['Date Valued'].dt.year
data.head()
```

```python
data.head(10)
```

```python
data.to_csv('Clean_Nav.csv', index=False)
```

```python
import matplotlib.pyplot as plt
```

```python
for scheme_name in data['Scheme Name'].unique():
    scheme_data = data[data['Scheme Name'] == scheme_name]
    scheme_data.plot(x='Date Valued', y='Sale Price/Unit')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title(f'Sale Price Performance Over Time - {scheme_name}')
    plt.legend(loc='best')
    plt.show()
```
